```python
A = [17,22,19]
B = [10,20,11]
C = [5,12,9]

A = np.array(A)
B = np.array(B)
C = np.array(C)

print(A, B, C)
```

```python
from numpy.core.multiarray import dot
# 1.
hasil1 = A + B
print('1. ', hasil1)

# 2.
hasil2 = B - C
print('2. ', hasil2)

# 3.
hasil3 = np.dot(A,C)
print('3. ', hasil3)

# 4.
hasil4 = A * B
print('4. ', hasil4)

# 5.
hasil5 = np.linalg.norm(A)
print('5. norm A = ', hasil5)

# 6.
hasil6 = np.arccos(np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B)))
sudut6 = np.degrees(hasil6)
print('6. ', hasil6, ', sudut = ', sudut6)
```

```python
U = [2,5]
V = [3,1]

U = np.array(U)
V = np.array(V)
```

```python
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
origin = [0], [0]  # Titik asal
plt.quiver(*origin, U[0], U[1], color='orange', angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, V[0], V[1], color='purple', angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 4)
plt.ylim(0, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot Vektor')
plt.grid()
plt.legend(['U', 'V'])
plt.show()
```

```python
# Buat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definisikan titik awal (0, 0, 0) dan arah vektor U dan V
origin = [0], [0], [0]
ax.quiver(*origin, U[0], U[1], 0, color='red')
ax.quiver(*origin, V[0], V[1], 0, color='blue')

# Atur batas sumbu x, y, dan z
ax.set_xlim([0, max(U[0], V[0]) + 1])
ax.set_ylim([0, max(U[1], V[1]) + 1])
ax.set_zlim([0, 1])

# Tampilkan plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Plot Vektor dalam Bidang 3D')
plt.show()
```
